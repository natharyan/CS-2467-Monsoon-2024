```python
import cv2
import matplotlib.pyplot as plt

org = cv2.imread('../../images/org.jpg')
plt.imshow(cv2.cvtColor(org,cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(cv2.cvtColor(org,cv2.COLOR_BGR2GRAY),cv2.COLOR_BGR2RGB))
plt.show()
```

```python
grayorg = cv2.cvtColor(org,cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(grayorg,cv2.COLOR_BGR2RGB))
plt.show()

sharpened = grayorg + (grayorg- cv2.GaussianBlur(grayorg,(15,15),0))
plt.imshow(cv2.cvtColor(sharpened,cv2.COLOR_BGR2RGB))
plt.show()
```
