```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
def convolution(x,h):
    lengthX = np.size(x)
    lengthH = np.size(h)
    y = np.zeros(lengthX + lengthH - 1)
    # for m in np.arange(lengthX):
    #     for n in np.arange(lengthH):
            # y[m + n] = y[m + n] + x[m]*h[n]
    for n in np.arange(lengthX):
        for k in np.arange(lengthH):
            y[n] = y[n] + x[k]*h[n-k]
        # y[n:n+lengthX] = y[n:n+lengthX] + x*h[n]
    return y
```

```python
t = np.linspace(0.15, 0.35, 500, endpoint=False)

plt.ylim(-0.5, 2.5)
sig1 = signal.square(2 * np.pi * 5.99 * t+4.4)+1
sig1 = sig1/2
plt.plot(sig1)
plt.grid('on')
```

```python
sigt = np.zeros_like(sig1)
sigt[150:350] = signal.triang(400)[200:]
plt.plot(sigt)
plt.grid('on')
```
